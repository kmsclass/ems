```python
#합계 컬럼을 총발전량 컬럼으로 변경하기
ndf = ndf.rename(columns={"합계":"총발전량"})
ndf.info()
```

```python
#전년도발전량 컬럼 추가
ndf["전년도발전량"]=ndf["총발전량"].shift(1)
```

```python
#증감율 컬럼 추가하기 : (현재-전년도)/전년도 * 100 => 퍼센트 단위
#                    = (현재/전년도 - 1) * 100
ndf["증감율"]=((ndf["총발전량"] / ndf["전년도발전량"]) - 1) * 100
```

```python
#연합 그래프 작성
#시각화 패키지 사용
import matplotlib.pyplot as plt
plt.style.available #사용 가능한 style 목록
plt.style.use("classic")
#한글폰트 설정
plt.rc("font",family="Malgun Gothic")
#음수표시 : -로 설정
plt.rcParams["axes.unicode_minus"]=False
```

```python
'''
ax1 : 막대그래프 영역. pandas.plot 함수로 그래프 작성
ndf[['수력','화력']] : 수력,화력데이터로만 그래프 작성
figsize=(20, 10) : 그래프 크기
width=0.7 : 막대그래프 넓이
stacked=False : 수력,화력의 막대를 따로따로 표시. 기본값
'''
ax1 = ndf[['수력','화력']].plot(kind='bar', figsize=(20, 10), \
                           width=0.7,stacked=False )  
'''
ax2 : 선그래프의 영역. plt 패키지로 그래프 작성.
'''
ax2 = ax1.twinx() #ax1,ax2 그래프가 같은 영역에서 출력
'''
ndf.index : x축의 데이터.
ndf.증감율 : y축의 데이터
ls='--' : 선의 종류 (-:실선, --:댓쉬선)
marker='o' : 선의 마커 표시
markersize=5 : 마커의 크기
color='green' : 선의 색상
label='전년대비 증감율(%)' : 범례에 표시될 데이터 설정
'''
ax2.plot(ndf.index, ndf.증감율, ls='-', marker='o', markersize=5, 
        color='green', label='전년대비 증감율(%)') #선그래프
ax1.set_ylim(0,200) #ax1 영역의 y축의 값의 범위 설정
ax2.set_ylim(-50,50) #ax2 영역의 y축의 값의 범위 설정
ax1.set_xlabel("년도",size=20) #x축의 라벨
ax1.set_ylabel("발전량(억 KWh)") #y축의 라벨
ax2.set_ylabel("전년 대비 증감율(%)") #y축의 라벨
plt.title("북한 전력 발전량(1990~2014)",size=30)
ax1.legend(loc="upper left") #범례. loc=: 범례의 위치. 왼쪽 위
ax2.legend(loc="upper right")
#이미지 저장하기
plt.savefig("북한전력량.png",dpi=400,bbox_inches="tight")
plt.show()
```

```python
#박스그래프:box 그래프 
# 두개의 그래프로 출력하기
#그래프 출력을 위한 영역 생성
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1) #1행2열 1번째 영역
ax2 = fig.add_subplot(1,2,2) #1행2열 2번째 영역
'''
  x=[usa인연비데이터목록,japan인연비데이터목록,europe인연비데이터목록]
  vert=False : 수평박스플롯으로 설정
  sym : 이상치 표현 방식
       r* : red, *
            r:red, b:blue,...
            *:별, s:사각형,.:점, o:원, d:다아몬드, +:십자가
'''
box1 = ax1.boxplot(x=[mpg[mpg["origin"]=='usa']['mpg'],
                      mpg[mpg["origin"]=='japan']['mpg'],
                      mpg[mpg["origin"]=='europe']['mpg']],
                      labels=["USA","JAPAN","EU"],sym="r+")
box2 = ax2.boxplot(x=[mpg[mpg["origin"]=='usa']['mpg'],
                      mpg[mpg["origin"]=='japan']['mpg'],
                      mpg[mpg["origin"]=='europe']['mpg']],
                      labels=["USA","JAPAN","EU"],vert=False)
ax1.set_title("제조국별 연비 분포(수직박스플롯)")
ax2.set_title("제조국별 연비 분포(수평박스플롯)")
```

```python

seoul_map = folium.Map(location=[37.55,126.98],zoom_start=13)
df = pd.read_excel("data/서울지역 대학교 위치.xlsx",index_col=0)
```

```python
#원형 마커 추가하기
for name,lat,lng in zip(df.index,df.위도,df.경도) :
    folium.CircleMarker([lat,lng],popup=name,tooltip=name,
                        radius=10,   #원의 반지름
                        color='brown', #원둘레 색상
                        fill=True,     #원 내부 채움
                        fill_color='coral', #원 내부의 색상
                        fill_opacity=0.5 # 원 내부의 투명도(?). 
                        ).add_to(seoul_map)
seoul_map.save("seoul3.html")
```

```python
#마커 내부의 아이콘 설정하기
#icon :home,flag,bookmark,star
for name,lat,lng in zip(df.index,df.위도,df.경도) :
    folium.Marker([lat,lng],popup=name,tooltip=name,
                        icon=folium.Icon(color="red",icon="bookmark")
                        ).add_to(seoul_map)
seoul_map.save("seoul4.html")
```
