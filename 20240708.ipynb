```python
df.info()
#북한지역의 발전량 데이터만 ndf 저장하기
ndf = df.loc[5:]
ndf
```

```python
#전력량 (억㎾h) 컬럼 제거
#del ndf["전력량 (억㎾h)"]
ndf = ndf.drop("전력량 (억㎾h)",axis=1)
ndf.info()
```

```python
#발전 전력별 데이터를 인덱스로 변경하기 : set_index
ndf.set_index("발전 전력별",inplace=True)
ndf.info()
```

```python
#ndf 데이터를 전치행열로 변경하기
ndf = ndf.T
ndf.info()
```

```python
#합계 컬럼을 총발전량 컬럼으로 변경하기
ndf = ndf.rename(columns={"합계":"총발전량"})
ndf.info()
```

```python
#전년도발전량 컬럼 추가
ndf["전년도발전량"]=ndf["총발전량"].shift(1)
```

```python
ndf
```

```python
#증감율 컬럼 추가하기 : (현재-전년도)/전년도 * 100 => 퍼센트 단위
#                    = (현재/전년도 - 1) * 100
ndf["증감율"]=((ndf["총발전량"] / ndf["전년도발전량"]) - 1) * 100
```
